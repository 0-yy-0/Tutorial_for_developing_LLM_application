# Fast api 进行前后端分离 💬

目前我们已经完成了基本的可视化页面，并且可以实现对应的功能。

为了方便整个项目的管理，现有的项目通常采用前后端分离的方式搭建，前后端数据通过 json 的格式进行传输。


FastAPI 是一个用于构建 API 的现代、快速（高性能）的 web 框架，非常方便用于搭建我们的前后端分离的应用。

在前面的章节中，我们已经搭建好 gradio 前端界面了，我们可以直接将其与 Fastapi 的 APP 绑定。

```python
import gradio as gr
from fastapi import FastAPI

app = FastAPI() # 创建 FastApi app
app = gr.mount_gradio_app(app, demo, path="/") # 将前面章节的前端页面和 FastApi 绑定
```

运行下面的命令就可以启动我们的 APP 了。

```python
uvicorn app:app 
```
